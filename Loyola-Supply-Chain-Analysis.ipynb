```python
for i in range(len(Q1)):
    if Q1['Mileage'][i]==0:
        rpm[i]=0
    else:
        try:
            rpm[i]= Q1['Revenue'][i]/Q1['Mileage'][i]
        except TypeError:
            rpm[i]=0
```

```python
Q1['RPM'] = rpm
```
